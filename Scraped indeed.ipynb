```python
import json
import html_parsing
from pprint import pprint

with open('data/resumes2.json') as data_file:
    data = json.load(data_file)
```

```python
resumes = []
    
def simplify(attr):
    if len(attr) == 0:
        return None
    return html_parsing.strip_tags(attr[0])
    
for resume in data:
    entries = []
    resumes.append(entries)
    for entry in resume['entries']:
        company = simplify(entry['company'])
        dates = simplify(entry['dates'])
        title = simplify(entry['title'])
        if company is not None and dates is not None:
            entries.append((company, dates))
        
len(resumes)
```

```python
pprint(resumes)
```
